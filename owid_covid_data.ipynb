```python
#Lets find the continent most affected by COVID 19 and the least one and their number.
#1. South America was the most affected continent with total deaths of 341,631,554 people
#2. Australia / Oceania was the least affected continent with total deaths of 743,222 people.
total = data.groupby('continent').sum()
groupedTotal = total.sort_values(by="total_deaths" , ascending = False)
groupedTotal
```
